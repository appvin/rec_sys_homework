```python
import pandas as pd
import numpy as np
from scipy.sparse import lil_matrix
from scipy.spatial import distance
import math
from sklearn.metrics.pairwise import cosine_similarity
from zscore import *

#u,i,rui,pui
def RMSE(records):
    rui=records[2]   
    pui=records[3]
    return math.sqrt(sum((rui-pui).dot(rui-pui))/float(len(records)))

def ItemSimilarity(train):
    nUsers = max(train[0])+1
    nItems = max(train[1])+1
    R = lil_matrix((nUsers, nItems))
    for idx,row in train.iterrows():
        R[row[0],row[1]]=row[2]
    S = cosine_similarity(R.T)
    W=dict()
    for u in range(1,nItems):
        for i in range(1,nItems):
            if u not in W:
                W.update({u:{i: S[u][i]}})
            else:
                W[u][i]=S[u][i]
    return W

def ratingPredict(u,i,W,train_dict):
    num=0
    sumrating=0
    sumsim=0
    if i not in W:
        print(['lost i: ',i])
        return 3
    for simi,w in W[i]:
        if simi not in train_dict[u]:
            continue
        num+=1
        sumrating+=train_dict[u][simi]*w
        sumsim+=w
        #if sum==10:
            #break
    if(sumsim==0):
        #print([u,i])
        return -1
    return sumrating/sumsim

def itemcf_test(train,test):
    #[train,meanr,stdr]=zscore.h_item(train)
    train_dict=dict()
    for index,row in train.iterrows():
        u=row[0]
        i=row[1]
        rating=row[2]
        if u not in train_dict:
            train_dict.update({u:{i: rating}})
        else:
            train_dict[u][i]=rating  
    print('train...')
    W=ItemSimilarity(train)
    for i,w in W.items():
        W[i]=sorted(W[i].items(),  key=lambda d:d[1], reverse = True)
    result=pd.DataFrame(dtype=float)
    result[0]=test[0]
    result[1]=test[1]
    result[2]=test[2]
    result[3]=0.0
    for index,row in result.iterrows():
        result.loc[index,3]=ratingPredict(row[0],row[1],W,train_dict)
    #result=zscore._h_item(result,meanr,stdr)
    #print(result)
    #print(RMSE(result))
    return result

def itemcf(train):
    #[train,meanr,stdr]=zscore.h_item(train)
    train_dict=dict()
    for index,row in train.iterrows():
        u=row[0]
        i=row[1]
        rating=row[2]
        if u not in train_dict:
            train_dict.update({u:{i: rating}})
        else:
            train_dict[u][i]=rating  
    print('train...')
    W=ItemSimilarity(train)
    for i,w in W.items():
        W[i]=sorted(W[i].items(),  key=lambda d:d[1], reverse = True)
    result=pd.DataFrame()
    nUsers = max(train[0])+1
    nItems = max(train[1])+1
    for u in range(1,nUsers):
        for i in range(1,nItems):
            eui=ratingPredict(u,i,W,train_dict)
            if eui<0:
                continue
            row=[u,i,eui]
            if result.empty:
                result[0]=row
            else:
                result=result.append(row,ignore_index=True)
            if result.size%10000==0:
                print(result.size)
    return result
    
if __name__ == "__main__": 
    train = pd.read_table('../ml-100k/ua.base', header=None)
    test = pd.read_table('../ml-100k/ua.test', header=None) 
    print(itemcf(train))
```
