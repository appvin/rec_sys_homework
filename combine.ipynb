```python
import math
import numpy as np
from sklearn import linear_model
import pandas as pd

#u,i,rui,pui
def RMSE(records):
    rui=records[2]   
    pui=records[3]
    return math.sqrt(sum((rui-pui).dot(rui-pui))/float(len(records)))

if __name__ == "__main__": 
    train1=pd.read_csv('CF/usercf-train.csv')
    train2=pd.read_csv('CF/itemcf-train.csv')
    test1=pd.read_csv('CF/usercf-test.csv')
    test2=pd.read_csv('CF/itemcf-test.csv')
    #print(train1)
    train1.columns=['u','i','r','e1']
    train2.columns=['u','i','r','e2']
    train=pd.merge(train1,train2,on=['u','i','r'],how='inner')
    #print(x)
    regr = linear_model.LinearRegression()
    x=pd.DataFrame()
    x['e1']=train['e1']
    x['e2']=train['e2']
    y=train['r']
    regr.fit(x, y)
    test1.columns=['u','i','r','e1']
    test2.columns=['u','i','r','e2']
    test=pd.merge(test1,test2,on=['u','i','r'],how='inner')
    result=pd.DataFrame()
    result[0]=test['u']
    result[1]=test['i']
    result[2]=test['r']
    testx=pd.DataFrame()
    testx['e1']=test['e1']
    testx['e2']=test['e2']
    result[3]=regr.predict(testx)
    print(result)
    print(RMSE(result))
```
