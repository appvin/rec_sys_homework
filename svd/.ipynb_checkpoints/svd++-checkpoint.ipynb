```python
import math
import random
import pandas as pd
from scipy.sparse import csr_matrix

#u,i,rui,pui
def RMSE(records):
    rui=records[2]   
    pui=records[3]
    return math.sqrt(sum((rui-pui).dot(rui-pui))/float(len(records)))

def InitBiasLFM(train,F):
    p=dict()
    q=dict()
    bu=dict()
    bi=dict()
    mu=mean(train[2])
    for index,row in train.iterrows():
        u=row[0]
        i=row[1]
        bu[u]=0
        bi[i]=0
        if u not in p:
            p[u]=[random.random()/math.sqrt(F) for x in range(0,F)]
        if i not in q:
            q[i]=[random.random()/math.sqrt(F) for x in range(0,F)]
    return [p,q,bu,bi,mu]

def Predict(u,i,p,q,bu,bi,mu):
    if u not in p:
        print(['lost u',u])
        return mu
    if i not in q:
        print(['lost i',i])
        return mu
    ret=mu+bu[u]+bi[i]
    ret+=sum(p[u][f]*q[i][f] for f in range(0,len(p[u])))
    return ret

def LearningBiasLFM(train,F,n,alpha,lamda):
    [p,q,bu,bi,mu,y]=InitBiasLFM(train,F)
    z=dict()
    print('learning:')
    for step in range(0,n):
        print(step)
        for index,row in train.iterrows():
            u=row[0]
            i=row[1]
            rui=row[2]
            z[u]=p[u]
            pui=Predict(u,i,p,q,bu,bi,mu)
            eui=rui-pui
            bu[u] += alpha * (eui - lamda * bu[u])
            bi[i] += alpha * (eui - lamda * bi[i]) 
            for k in range(0,F):
                p[u][k]+=alpha*(q[i][k]*eui-lamda*p[u][k])
                q[i][k]+=alpha*(p[u][k]*eui-lamda*q[i][k])
        alpha*=0.9
    return [bu,bi,mu,p,q]

if __name__ == "__main__":  
    train = pd.read_table('ua.base', header=None)
    test = pd.read_table('ua.test', header=None)
    train=train.pivot_table(index=[0,1],values=[2])
    #print(train)
    for i in train.index.levels[0]: 
        for index,row in train.loc[i].iterrows():
            print([index,row[2]])
        break
    '''[bu,bi,mu,p,q]=LearningBiasLFM(train,1,1,0.02,0.01)
    result=pd.DataFrame(dtype=float)
    result[0]=test[0]
    result[1]=test[1]
    result[2]=test[2]
    result[3]=0.0
    for index,row in result.iterrows():
        result.loc[index,3]=Predict(row[0],row[1],p,q,bu,bi,mu)
    print(result)
    print(RMSE(result))'''
```
