```python
import time
import heapq

top_N=5

all_t=time.time()
df=pd.read_csv('10000-samples.dat',\
               sep='\s+',names=['User','Item','Ratio','Timestamp'])

# Index User and Item via number
print('num : ',df.shape[0])
print('count : user item')
user2index = dict()
item2index = dict()
user_n=0
item_n=0

for user in df['User']:
    if user not in user2index:
        user2index[user]=user_n
        user_n+=1
for item in df['Item']:
    if item not in item2index:
        item2index[item]=item_n
        item_n+=1

print (user_n,item_n)
print(time.time()-all_t,'s')


#init matrix
print('init user_item matrix')
t0=time.time()
user_item=dok_matrix((user_n,item_n))
for i,record in df.iterrows():
    if(i>10000):
        break
    user_item[user2index[record['User']],item2index[record['Item']]]=1
print(time.time()-t0,'s','end')

#init item_matrix
print('init item_matrix')
t0=time.time()
item_matrix=dok_matrix((item_n,item_n))
for i in range(0,user_n):
    nonzero=user_item[i].nonzero()[1]
    item_num=len(nonzero)
    for j in range(0,item_num):
        for k in range(j+1,item_num):
            item_matrix[nonzero[j],nonzero[k]]=1
            item_matrix[nonzero[k],nonzero[j]]=1
print(time.time()-t0,'s','end')

#norm
print('norm')
t0=time.time()

print(time.time()-t0,'s','end')

#cal result,every col is a user's result
print('cal result')
t0=time.time()
result_matrix=item_matrix*user_item.T
for i in range(0,user_n):
    nonzero=result_matrix.getcol(i)
    print(heapq.nlargest(top_N,nonzero))
    break
print(time.time()-t0,'s','end')

print(time.time()-all_t,'s','all end')
```
