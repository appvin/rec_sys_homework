```python
import math
import random
import pandas as pd
from scipy.sparse import csr_matrix

#u,i,rui,pui
def RMSE(records):
    rui=records[2]   
    pui=records[3]
    return math.sqrt(sum((rui-pui).dot(rui-pui))/float(len(records)))

def InitSVDpp(train,F):
    p=dict()
    q=dict()
    bu=dict()
    bi=dict()
    y=dict()
    mu=mean(train[2])
    for index,row in train.iterrows():
        u=row[0]
        i=row[1]
        bu[u]=0
        bi[i]=0
        if u not in p:
            p[u]=[random.random()/math.sqrt(F) for x in range(0,F)]
        if i not in q:
            q[i]=[random.random()/math.sqrt(F) for x in range(0,F)]
        if i not in y:
            y[i]=[random.random()/math.sqrt(F) for x in range(0,F)]
    return [p,q,bu,bi,mu,y]

def Predict(u,i,p,q,bu,bi,mu,z):
    if u not in p:
        print(['lost u',u])
        return mu
    if i not in q:
        print(['lost i',i])
        return mu
    ret=mu+bu[u]+bi[i]
    ret+=sum(q[i][f]*z[u][f] for f in range(0,len(p[u])))
    return ret

def LearningSVDpp(train,F,n,alpha,lamda):
    [p,q,bu,bi,mu,y]=InitSVDpp(train,F)
    z=dict()
    train=train.pivot_table(index=[0,1],values=[2])
    print('learning:')
    for step in range(0,n):
        print(step)
        for u in train.index.levels[0]:
            z[u]=p[u]
            items=train.loc[u]
            ru=1/math.sqrt(1.0*len(items))
            for i,row in items.iterrows():
                rui=row[2]
                for f in range(0,F):
                    z[u][f]+=y[i][f]*ru
            sumall=[0 for i in range(0,F)]
            for i,row in items.iterrows():
                rui=row[2]
                pui=Predict(u,i,p,q,bu,bi,mu,z)
                eui=rui-pui
                bu[u]+=alpha*(eui-lamda*bu[u])
                bi[i]+=alpha*(eui-lamda*bi[i])
                for f in range(0,F):
                    sumall[f]+=q[i][f]*eui*ru
                    p[u][f]+=alpha*(q[i][f]*eui-lamda*p[u][f])
                    q[i][f]+=alpha*(z[u][f]*eui-lamda*q[i][f])
            for i,row in items.iterrows():
                for f in range(0,F):
                    y[i][f]+=alpha*(sumall[f]-lamda*y[i][f])
        alpha*=0.9
    return [bu,bi,mu,p,q,z]

if __name__ == "__main__":  
    train = pd.read_table('ua.base', header=None)
    test = pd.read_table('ua.test', header=None)    
    [bu,bi,mu,p,q,z]=LearningSVDpp(train,1,1,0.02,0.01)
    result=pd.DataFrame(dtype=float)
    result[0]=test[0]
    result[1]=test[1]
    result[2]=test[2]
    result[3]=0.0
    for index,row in result.iterrows():
        result.loc[index,3]=Predict(row[0],row[1],p,q,bu,bi,mu,z)
    print(result)
    print(RMSE(result))
```
