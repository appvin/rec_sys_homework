```python
import pandas as pd

df=pd.read_csv('2008.csv')
print(df)
```

```python
print(df[df['Cancelled']>1])
```

```python
re=df[df['Cancelled']!=1]
re=re[re['Cancelled']!=0]
print(re)
```

```python
rr=pd.DataFrame()
rr['UniqueCarrier']=df['UniqueCarrier']
rr['Cancelled']=df['Cancelled']
gr=rr.groupby('UniqueCarrier')['Cancelled'].sum()
print(gr)
```

```python
grs=gr.sort_values(ascending=False)
print(grs)
```
