```python
import math
import random
import pandas as pd
from scipy.sparse import csr_matrix

#u,i,rui,pui
def RMSE(records):
    rui=records.vote   
    pui=records.predict
    return math.sqrt(sum((rui-pui).dot(rui-pui))/float(len(records)))

def AddToDict(dic,i,j,v=-1):
    if v==-1:
        v=j
        if i not in dic:
            dic[i]=v
        else:
            dic[i]+=v
    else:
        if i not in dic:
            dic.update({i:{j: v}})
        elif j not in dic[i]:
            dic[i][j]=v
        else:
            dic[i][j]+=v

class Cluster:
    def __init__(self,records):
        self.group = dict()
        
    def GetGroup(self, i):
        return 0

class UserVoteCluster(Cluster):
    def __init__(self, records): 
        Cluster.__init__(self, records)
        vote = dict() 
        count = dict() 
        for index,r in records.iterrows():
            if r.test != 0:
                continue            
            AddToDict(vote, r.user, r.vote)
            AddToDict(count, r.user, 1)
        k = 0
        for user, v in vote.items():
            ave = v / (count[user] * 1.0) 
            c = int(ave * 2) 
            self.group[user] = c
            
    def GetGroup(self, uid): 
        if uid not in self.group: 
            return -1
        else:
            return self.group[uid]

class ItemVoteCluster(Cluster):
    def __init__(self, records): 
        Cluster.__init__(self, records)
        vote = dict()
        count = dict() 
        for index,r in records.iterrows():
            if r.test != 0:
                continue          
            AddToDict(vote, r.i, r.vote) 
            AddToDict(count, r.i, 1)
        k = 0
        for i, v in vote.items(): 
            ave = v / (count[i] * 1.0)
            c = int(ave * 2)
            self.group[i] = c
        
    def GetGroup(self, i):
        if i not in self.group: 
            return -1
        else:
            return self.group[i]

def PredictAll(records, user_cluster, item_cluster):
    total = dict()
    count = dict() 
    meanall=mean(records[records['test']==0]['vote'])
    for index,r in records.iterrows():
        if r.test != 0: 
            continue
        gu = user_cluster.GetGroup(r.user)
        gi = item_cluster.GetGroup(r.i)
        AddToDict(total, gu, gi, r.vote) 
        AddToDict(count, gu, gi, 1) 
    for index,r in records.iterrows():
        if index%10000==0:
            print(index)
        gu = user_cluster.GetGroup(r.user) 
        gi = item_cluster.GetGroup(r.i)
        if gu not in total or gi not in total[gu]:
            average=meanall
        else:
            average = total[gu][gi] / (1.0 * count[gu][gi] + 1.0)
        records.loc[index,'predict']=average
    return records

def Predict(train, test, alpha,user_cluster, item_cluster): 
    total = dict() 
    count = dict()
    submean=mean(train['vote'])-mean(train['predict'])
    for index,record in train.iterrows():
        gu = user_cluster.GetGroup(record.user) 
        gi = item_cluster.GetGroup(record.i)
        AddToDict(total, gu, gi, record.vote - record.predict) 
        AddToDict(count, gu, gi, 1) 
    for index,record in test.iterrows():
        gu = user_cluster.GetGroup(record.user) 
        gi = item_cluster.GetGroup(record.i)
        if gu not in total or gi not in total[gu]:
            average=submean
        else:
            average = total[gu][gi] / (1.0 * count[gu][gi] + alpha) 
        record.predict += average
    for index,record in train.iterrows():
        gu = user_cluster.GetGroup(record.user) 
        gi = item_cluster.GetGroup(record.i)
        if gu not in total or gi not in total[gu]:
            average=submean
        else:
            average = total[gu][gi] / (1.0 * count[gu][gi] + alpha) 
        record.predict += average
    return [train,test]

def meanpredict_test(train,test):  
    train['test']=0
    test['test']=1
    records=pd.concat([train,test],ignore_index=True)
    records.columns=['user','i','vote','timestamp','test'] 
    records['predict']=0.0
    usercluster=UserVoteCluster(records)
    itemcluster=ItemVoteCluster(records)
    results=PredictAll(records,UserVoteCluster(records),ItemVoteCluster(records))
    tests=results[results.test==1]
    print(RMSE(tests))
    trains=results[results.test==0]
    alpha=0.9
    for i in range(0,10):
        [trains,tests]=Predict(trains,tests,alpha,usercluster,itemcluster)
        alpha*=0.9
    print(tests)
    print(RMSE(tests))
    
def meanpredict(train):
    train_dict=dict()
    for index,row in train.iterrows():
        u=row[0]
        i=row[1]
        rating=row[2]
        if u not in train_dict:
            train_dict.update({u:{i: rating}})
        else:
            train_dict[u][i]=rating   
    train['test']=0
    train.drop(3,axis=1)
    trian=pd.DataFrame()
    test=pd.DataFrame()
    nUsers = max(train[0])+1
    nItems = max(train[1])+1
    num=0
    for u in range(1,nUsers):
        for i in range(1,nItems):
            if u in train_dict and i in train_dict[u]:
                continue
            else:
                row=[u,i,0]
                if trian.empty:
                    trian[0]=row
                else:
                    trian=trian.append(row,ignore_index=True)
    test['test']=1
    records=pd.concat([train,test],ignore_index=True)
    records.columns=['user','i','vote','test'] 
    records['predict']=0.0
    usercluster=UserVoteCluster(records)
    itemcluster=ItemVoteCluster(records)
    results=PredictAll(records,UserVoteCluster(records),ItemVoteCluster(records))
    tests=results[results.test==1]
    print(RMSE(tests))
    trains=results[results.test==0]
    alpha=0.9
    for i in range(0,10):
        [trains,tests]=Predict(trains,tests,alpha,usercluster,itemcluster)
        alpha*=0.9
    print(tests)
    print(RMSE(tests))
    trains.drop('test',axis=1)
    tests.drop('test',axis=1)
    trains.columns=[1,2,3,4]
    tests.columns=[1,2,3,4]
    return [trains,tests]
    
if __name__ == "__main__": 
    train = pd.read_table('../ml-100k/ua.base', header=None)
    test = pd.read_table('../ml-100k/ua.test', header=None) 
    print(meanpredict(test))
```
